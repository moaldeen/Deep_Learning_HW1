```python

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

import torchvision
from torchvision import datasets, transforms

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

# Set up data transformations
data_transforms = transforms.Compose([
                        transforms.ToTensor()])

def load_dataset(training_batch_size, validation_batch_size):
    training_loader = torch.utils.data.DataLoader(
                            datasets.MNIST(root='./dataset', train=True, download=True, transform=data_transforms),
                            batch_size=training_batch_size, shuffle=True)

    validation_loader = torch.utils.data.DataLoader(
                            datasets.MNIST(root='./dataset', train=False, download=True, transform=data_transforms),
                            batch_size=validation_batch_size, shuffle=False)

    return (training_loader, validation_loader)

class NeuralNet(nn.Module):
    def __init__(self):
        super(NeuralNet, self).__init__()
        self.dense1 = nn.Linear(784, 18)
        self.dense2 = nn.Linear(18, 20)
        self.dense3 = nn.Linear(20, 10)

    def forward(self, x):
        x = F.relu(self.dense1(x))
        x = F.relu(self.dense2(x))
        x = self.dense3(x)
        return x

def optimize_model(network, training_set, optimizer):
    network.train()
    loss_history = []
    accuracy_history = []
    num_epochs = 40
    epoch_data=pd.DataFrame()

    for epoch in range(num_epochs):
        epoch_loss = 0
        total = 0
        correct = 0
        print("Epoch - ", epoch+1)

        for batch_idx, (inputs, labels) in enumerate(training_set):
            optimizer.zero_grad()
            inputs = inputs.view(-1, 784)
            outputs = network(inputs)

            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            epoch_loss += loss.item()

            total += labels.size(0)
            _, predicted = torch.max(outputs.data, 1)
            correct += (predicted == labels).sum().item()

        epoch_accuracy = (correct / total) * 100

        loss_history.append(epoch_loss)
        accuracy_history.append(round(epoch_accuracy, 3))
        print('Epoch Accuracy - ', epoch_accuracy)
        print('Rounded Accuracy - ', round(epoch_accuracy, 3))

        # Save model weights
        weights_df = pd.DataFrame()
        for name, param in network.named_parameters():
            if 'weight' in name:
                weights = torch.nn.utils.parameters_to_vector(param).detach().numpy()
                weights_df = pd.concat([weights_df, pd.DataFrame(weights).T], axis=1)

        epoch_data = pd.concat([epoch_data, weights_df], axis=0)

    return loss_history, accuracy_history, epoch_data

training_loader, validation_loader = load_dataset(1000, 1000)
criterion = nn.CrossEntropyLoss()
model_weights=pd.DataFrame()
all_accuracy = []
all_loss = []

for i in range(8):
    print('Iteration - ', i)
    net = NeuralNet()
    optimizer = torch.optim.Adam(net.parameters(), lr=0.01)
    losses, accuracies, weights = optimize_model(net, training_loader, optimizer)
    all_loss.extend(losses)
    all_accuracy.extend(accuracies)
    model_weights = model_weights.append(weights)

weights_array = np.array(model_weights)
print(model_weights.shape)
pca = PCA(n_components=2)
transformed_weights = pca.fit_transform(weights_array)
transformed_df = pd.DataFrame(transformed_weights, columns=['dim1', 'dim2'])
print(transformed_df.shape)
print(all_accuracy)
transformed_df["accuracy"] = all_accuracy
transformed_df["loss"] = all_loss
print(transformed_df)

selected_data = transformed_df.iloc[::3, :]
print(selected_data)

for index in range(selected_data.shape[0]):
    accuracy_marker = list(selected_data['accuracy'])[index]
    plt.scatter(selected_data['dim1'][index*3], selected_data['dim2'][index*3], marker=f'${accuracy_marker}$')
    plt.title("Neural Network Model")


```
